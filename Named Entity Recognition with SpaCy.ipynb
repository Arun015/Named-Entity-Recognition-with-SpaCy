```python
from bs4 import BeautifulSoup
import requests
import re
def url_to_string(url):
    res = requests.get(url)
    html = res.text
    soup = BeautifulSoup(html, 'html5lib')
    for script in soup(["script", "style", 'aside']):
        script.extract()
    return " ".join(re.split(r'[\n\t]+', soup.get_text()))
```

```python
ny_bb = url_to_string('https://www.timesnownews.com/india/article/china-has-never-recognised-arunachal-pradesh-beijing-provokes-india-amid-ongoing-border-dispute/648965')
```

```python
article = nlp(ny_bb)
len(article.ents)
```

```python
labels = [x.label_ for x in article.ents]
Counter(labels)
```

```python
items = [x.text for x in article.ents]
Counter(items).most_common(3)
```
